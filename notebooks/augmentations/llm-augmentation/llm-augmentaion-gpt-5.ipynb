````python
import os
import json
import glob
import pandas as pd
from tqdm import tqdm
from openai import OpenAI

client = OpenAI()

INPUT_FILE = r".\datasets\raw_datasets_processed\train.csv"
WORK_DIR = "batch_work"
OUTPUT_DIR = "judge_results"

os.makedirs(WORK_DIR, exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ---------------------------------------------------------------------------
# PROMPTS + Convert CSV → JSONL for Batch API
# ---------------------------------------------------------------------------

all_rows = []
request_id = 0

df = pd.read_csv(INPUT_FILE)

for idx, row in df.iterrows():
    if row["source"] == "ambari":
        if row["priority"] == "Blocker":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_BLOCKER_CLASS[0]["AMBARI"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_BLOCKER_CLASS[0]["AMBARI"]
        elif row["priority"] == "Critical":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_CRITICAL_CLASS[0]["AMBARI"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_CRITICAL_CLASS[0]["AMBARI"]
        elif row["priority"] == "Trivial":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_TRIVIAL_CLASS[0]["AMBARI"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_TRIVIAL_CLASS[0]["AMBARI"]
        elif row["priority"] == "Minor":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_MINOR_CLASS[0]["AMBARI"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_MINOR_CLASS[0]["AMBARI"]
        else:
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_MAJOR_CLASS[0]["AMBARI"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_MAJOR_CLASS[0]["AMBARI"]
    elif row["source"] == "camel":
        if row["priority"] == "Blocker":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_BLOCKER_CLASS[0]["CAMEL"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_BLOCKER_CLASS[0]["CAMEL"]
        elif row["priority"] == "Critical":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_CRITICAL_CLASS[0]["CAMEL"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_CRITICAL_CLASS[0]["CAMEL"]
        elif row["priority"] == "Trivial":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_TRIVIAL_CLASS[0]["CAMEL"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_TRIVIAL_CLASS[0]["CAMEL"]
        elif row["priority"] == "Minor":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_MINOR_CLASS[0]["CAMEL"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_MINOR_CLASS[0]["CAMEL"]
        else:
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_MAJOR_CLASS[0]["CAMEL"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_MAJOR_CLASS[0]["CAMEL"]
    elif row["source"] == "derby":
        if row["priority"] == "Blocker":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_BLOCKER_CLASS[0]["DERBY"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_BLOCKER_CLASS[0]["DERBY"]
        elif row["priority"] == "Critical":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_CRITICAL_CLASS[0]["DERBY"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_CRITICAL_CLASS[0]["DERBY"]
        elif row["priority"] == "Trivial":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_TRIVIAL_CLASS[0]["DERBY"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_TRIVIAL_CLASS[0]["DERBY"]
        elif row["priority"] == "Minor":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_MINOR_CLASS[0]["DERBY"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_MINOR_CLASS[0]["DERBY"]
        else:
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_MAJOR_CLASS[0]["DERBY"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_MAJOR_CLASS[0]["DERBY"]
    else:
        if row["priority"] == "Blocker":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_BLOCKER_CLASS[0]["WICKET"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_BLOCKER_CLASS[0]["WICKET"]
        elif row["priority"] == "Critical":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_CRITICAL_CLASS[0]["WICKET"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_CRITICAL_CLASS[0]["WICKET"]
        elif row["priority"] == "Trivial":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_TRIVIAL_CLASS[0]["WICKET"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_TRIVIAL_CLASS[0]["WICKET"]
        elif row["priority"] == "Minor":
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_MINOR_CLASS[0]["WICKET"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_MINOR_CLASS[0]["WICKET"]
        else:
            SYSTEM_PROMPT = SYSTEM_MESSAGE_FOR_MAJOR_CLASS[0]["WICKET"]
            USER_PROMPT_TEMPLATE = USER_MESSAGE_FOR_MAJOR_CLASS[0]["WICKET"]

    prompt = USER_PROMPT_TEMPLATE.format(
        num_samples=10,
        source=row["source"],
        summary=row["summary"],
        description=row["description"],
        priority=row["priority"]
        )

    custom_id = f"{row['source']}-{row['priority']}-{idx}"

    batch_item = {
        "custom_id": custom_id,
        "method": "POST",
        "url": "/v1/responses",
        "body": {
            "model": "gpt-5-2025-08-07",
            "temperature": 0.7,
            "max_output_tokens": 1024,
            "input": [
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user", "content": prompt}
            ]
        }
    }

    all_rows.append(batch_item)
    request_id += 1

jsonl_path = os.path.join(WORK_DIR, "judge_batch_llm_augmentation.jsonl")

with open(jsonl_path, "w", encoding="utf-8") as f:
    for item in all_rows:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print(f"Saved JSONL with {len(all_rows)} requests → {jsonl_path}")
````

````python
# ---------------------------------------------------------------------------
# STEP 2 — Upload batch file
# ---------------------------------------------------------------------------

batch_file = client.files.create(
    file=open(jsonl_path, "rb"),
    purpose="batch"
)

print("Uploaded file_id:", batch_file.id)
````

````python
# ---------------------------------------------------------------------------
# STEP 3 — Create batch job
# ---------------------------------------------------------------------------

batch_job = client.batches.create(
    input_file_id=batch_file.id,
    endpoint="/v1/responses",
    completion_window="24h"
)

print("Batch job created:", batch_job.id)
````

````python
import json
import pandas as pd

records = []
with open(r"batch_work\batch_output-llm-augmentation.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        item = json.loads(line)
        custom_id = item["custom_id"]
        
        text = None
        for output in item["response"]["body"]["output"]:
            for content in output.get("content", []):
                if content.get("type") == "output_text" and content.get("text", "").strip():
                    text = content["text"]
                    break
            if text:
                break
        
        if text is None:
            continue
        
        clean = text.strip()
        if clean.startswith("```json"):
            clean = clean[7:]
        if clean.endswith("```"):
            clean = clean[:-3]
        clean = clean.strip()
        
        try:
            samples = json.loads(clean)
            for sample in samples:
                records.append({
                    "id": custom_id,
                    "summary": sample.get("summary"),
                    "description": sample.get("description"),
                    "priority": sample.get("priority"),
                })
        except json.JSONDecodeError:
            print(f"Skipping {custom_id}: cannot parse JSON")

df = pd.DataFrame(records)
df
````
